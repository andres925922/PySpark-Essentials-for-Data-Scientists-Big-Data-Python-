### First things first..... import the pyspark sql functions library

Since we know we will be using it a lot.

```python
from pyspark.sql.functions import *
```

### 1. Select the Name and Position of each player in the dataframe

```python
fifa_19_df.select(["Name", "Position"]).limit(5).toPandas()
```

### 1.1 Display the same results from above sorted by the players names

```python
fifa_19_df.select(["Name", "Position"])                 \
        .orderBy(fifa_19_df["Name"].asc())             \
        .limit(5)                                       \
        .toPandas()
```

### 2. Select only the players who belong to a club begining with FC

```python
fifa_19_df.select(
    ["Name", "Club"]
).where(
    fifa_19_df["Club"].like(f"FC%")
).orderBy(fifa_19_df["Club"].asc()).limit(5).toPandas()
```

```python
oldest = fifa_19_df.select(
    ["Name", "Age"]
).orderBy(fifa_19_df["Age"].desc()).collect()

```

### 3. Who is the oldest player in the dataset and how old are they?

Display only the name and age of the oldest player.

```python
print(f"The oldest player is: {oldest[0][0]} with {oldest[0][1]} years old")
```

```python
# Another way to do this

fifa_19_df.select("Name", "Age").sort(desc("Age")).show(1)
```

### 4. Select only the following players from the dataframe:

 - L. Messi
 - Cristiano Ronaldo

```python
fifa_19_df.select(
    ["Name", "Club"]
).where(
    fifa_19_df["Name"].isin("L. Messi", "Cristiano Ronaldo")
).toPandas()
```

### 5. Can you select the first character from the Release Clause variable which indicates the currency used?

```python
fifa_19_df.select(
    "Name", "Release Clause", fifa_19_df["Release Clause"].substr(1,1).alias("Release Currency")
).show(5, False)
```

### 6. Can you select only the players who are over the age of 40?

```python
fifa_19_df.select("Name", "Age")                    \
            .where(fifa_19_df["Age"] > 40)          \
            .sort(fifa_19_df["Age"]. desc()).limit(10).toPandas()
```

```python
# Another way
fifa_19_df.select("Name", "Age").filter("Age>40")                 \
    .sort(fifa_19_df["Age"].desc())         \
    .limit(10).toPandas()
```
